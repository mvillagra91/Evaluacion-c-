# [dbo].[actualizar_producto]

```sql
/****** Object:  StoredProcedure [dbo].[actualizar_producto]    Script Date: 09/01/2025 13:41:06 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Procedimiento para actualizar un producto
CREATE PROCEDURE [dbo].[actualizar_producto](
    @id INT, 
    @nombre VARCHAR(20), 
    @descripcion VARCHAR(100), 
    @categoria VARCHAR(20)
)
AS
BEGIN
    BEGIN TRY
        UPDATE producto 
        SET name = @nombre, 
            description = @descripcion, 
            category = @categoria
        WHERE id_product = @id;
    END TRY
    BEGIN CATCH
        PRINT 'Error al actualizar el producto.';
    END CATCH
END;
GO
```

# [dbo].[producto_eliminar]

```sql
/****** Object:  StoredProcedure [dbo].[producto_eliminar]    Script Date: 09/01/2025 13:41:06 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Procedimiento para eliminar un producto (corregido el tipo de dato)
CREATE PROCEDURE [dbo].[producto_eliminar](
    @identificador_producto INT
)
AS
BEGIN
    BEGIN TRY
        DELETE FROM producto 
        WHERE id_product = @identificador_producto;
    END TRY
    BEGIN CATCH
        PRINT 'Error al eliminar el producto.';
    END CATCH
END;

GO
```

# [dbo].[producto_listar]

```sql
/****** Object:  StoredProcedure [dbo].[producto_listar]    Script Date: 09/01/2025 13:41:06 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Procedimiento para listar todos los productos
CREATE PROCEDURE [dbo].[producto_listar]
AS
BEGIN
    -- Selecciona todos los productos, ordenados por el ID del producto
    SELECT * 
    FROM producto
    ORDER BY id_product ASC; -- Puedes cambiar el orden si lo deseas
END;

GO
```

# [dbo].[registrar_producto]

```sql
/****** Object:  StoredProcedure [dbo].[registrar_producto]    Script Date: 09/01/2025 13:41:06 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[registrar_producto](
    @id INT,
    @nombre VARCHAR(20), 
    @descripcion VARCHAR(100), 
    @categoria VARCHAR(20)
)
AS
BEGIN
    BEGIN TRY
        -- Habilitar IDENTITY_INSERT solo si es necesario
        SET IDENTITY_INSERT [dbo].producto ON;

        INSERT INTO [dbo].[producto] ([id_product], [name], [description], [category])
        VALUES (@id, @nombre, @descripcion, @categoria);

        -- Deshabilitar IDENTITY_INSERT una vez usado
        SET IDENTITY_INSERT [dbo].producto OFF;
        
        PRINT 'Producto registrado exitosamente.';
    END TRY
    BEGIN CATCH
        -- Deshabilitar IDENTITY_INSERT en caso de error y mostrar el error
        IF @@TRANCOUNT > 0
            ROLLBACK TRANSACTION;

        IF OBJECTPROPERTY(OBJECT_ID('[dbo].producto'), 'TableHasIdentity') = 1
            SET IDENTITY_INSERT [dbo].producto OFF;

        PRINT 'Error al registrar el producto: ' + ERROR_MESSAGE();
    END CATCH
END;
GO
```
